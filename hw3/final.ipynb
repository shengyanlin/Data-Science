```python
import pandas as pd
```

```python
GCN = pd.read_csv('output.csv')
GCNH = pd.read_csv("GCNH.csv")
SSP = pd.read_csv("SSP1.csv")
```

```python
#combine "Predict" colunmn of GCN, GCNH, SSP named final
final = pd.concat([GCN['Predict'], GCNH['Predict'], SSP['Predict']], axis=1)
final.columns = ['GCN', 'GCNH', 'SSP']
```

```python
final
```

```python
#make final a str type
final = final.astype(str)
```

```python
cnt = 0
#if element is not all the same in the same row, then print the row and the index
 #make a new column named "result"
final['result'] = -1
for i in range(len(GCN)):   
    #1.If element in a row are all the same, then add that element to the "result"
    if final.iloc[i,0] == final.iloc[i,1] == final.iloc[i,2]:
        final.iloc[i,3] = final.iloc[i,0]
        print(i,"all the same")
    #2.0 == 1 != 2, 
    elif final.iloc[i,0] == final.iloc[i,1] != final.iloc[i,2]:
        final.iloc[i,3] = final.iloc[i,0]
        print(i,"0 == 1 != 2")
    #3.0 == 2 != 1
    elif final.iloc[i,0] == final.iloc[i,2] != final.iloc[i,1]:
        final.iloc[i,3] = final.iloc[i,0]
        print(i,"0 == 2 != 1")
    #4.1 == 0 != 2
    elif final.iloc[i,1] == final.iloc[i,2] != final.iloc[i,0]:
        final.iloc[i,3] = final.iloc[i,1]
        print(i,"1 == 2 != 0")
    #5.All different => 1
    else:
        print(i,"all different")
        final.iloc[i,3] = 1
        cnt+=1
print(cnt)
```

```python
final.to_csv("final.csv", index=False)
```

```python
#make a dataframe named "final" to store the result with "Id", "Predict" columns
finalout = pd.DataFrame(columns=['Id', 'Predict'])
#id is the index of the row, predict is the result of the row
finalout['Id'] = GCN['Id']
finalout['Predict'] = final['result']
finalout.to_csv("final.csv", index=False)
```
